## 17. Crie uma nova coluna chamada: “dormitory_type”
    - Se o valor da coluna “bedrooms” for igual à 1 => ‘studio’
    - Se o valor da coluna “bedrooms” for igual a 2 => ‘apartament’
    - Se o valor da coluna “bedrooms” for maior que 2 => ‘house’

```python
#Criar uma coluna nova, preenchida com NA
#Percorrer todas as linhas do conjunto de dados. Para cada linha, comparar a coluna "bedrooms"
#De acordo com a comparação, substituir o NA pelo dado valor

df2['dormitory_type'] = 'NA'
for i in range( len( df ) ):
    if df2.loc[i, 'bedrooms'] == 1:
        df2.loc[i, 'dormitory_type'] = 'studio'
        
    elif df2.loc[i, 'bedrooms'] == 2:
        df2.loc[i, 'dormitory_type'] = 'apartment'
        
    elif df2.loc[i, 'bedrooms'] > 2: 
        df2.loc[i, 'dormitory_type'] = 'house'
```

## 18. Crie uma nova coluna chamada: “condition_type”
    - Se o valor da coluna “condition” for menor ou igual à 2 => ‘bad’
    - Se o valor da coluna “condition” for igual à 3 ou 4 => ‘regular’
    - Se o valor da coluna “condition” for igual à 5 => ‘good’

```python
#Usar a função apply junto com a lambda para ter acesso a cada linha.
#Em cada linha, comparar a coluna "condition" com a condição dada.
#Aplicar a condição

df2['condition'] = df2['condition'].astype( int )
df2['conditional_type'] = df['condition'].apply( lambda x: 'bad' if x <= 2 else 'regular' if (x == 3) | (x == 4) else 'good' )
```
